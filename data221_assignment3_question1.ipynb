```python
#Import needed library
import pandas as pd
```

```python
#Loads dataset and gets wanted variable
crime_df = pd.read_csv("crime.csv")
violent_crimes = crime_df["ViolentCrimesPerPop"]
```

```python
#Calculate all needed statistics
mean_value = violent_crimes.mean()
median_value = violent_crimes.median()
standard_deviation = violent_crimes.std()
minimum_value = violent_crimes.min()
maximum_value = violent_crimes.max()
```

```python
#Print all statistics
print(f"Mean: {mean_value}")
print(f"Median: {median_value}")
print(f"Standard deviation: {standard_deviation}")
print(f"Minimum: {minimum_value}")
print(f"Maximum: {maximum_value}")
```
