```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Loads dataset and gets wanted variable
crime_df = pd.read_csv("crime.csv")
violent_crimes = crime_df["ViolentCrimesPerPop"]
```

```python
#Create Histogram
plt.figure(figsize=(10, 6))
plt.hist(violent_crimes, bins=20, edgecolor="black", color="skyblue")
plt.title("Distribution of Violent Crimes Per Population")
plt.xlabel("Violent Crimes Per Population")
plt.ylabel("Frequency")
plt.grid(axis="y")
plt.show()
```

```python
# Create box plot
plt.figure(figsize=(10, 6))
plt.boxplot(violent_crimes, vert = False)
plt.title("Box Plot of Violent Crimes Per Population")
plt.xlabel("Violent Crimes Per Population")
plt.grid(axis="x")
plt.show()
```

```python
'''
The histogram shows that the distribution of the data is skewed to the right. It shows that that there are a greater amount of communities with lower crime per population. On the flip-side there is fewer amount of communities with higher crime, however there is a relativity high amount of communities with a 1% ViolentCrimesPerPop at the right side of the histogram.

The boxplot shows the median that is represented by the line in the middle of the boxplot and it reinforces the idea of the distribution being right skewed since the median is closed to the 1Q(1st quartile) rather than the 3Q(3rd quartile).
This boxplot shows that there is not outliers in the data, since there are no points outside the whiskers.
'''
```
